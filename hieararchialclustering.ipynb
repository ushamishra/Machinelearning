```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# 1. Sample dataset with IDs
data = {
    'Customer_ID': ['Cust_A','Cust_B','Cust_C','Cust_D','Cust_E','Cust_F','Cust_G','Cust_H'],
    'Salary': [25000, 40000, 60000, 80000, 100000, 120000, 150000, 200000],
    'Spending_Score': [20, 35, 40, 60, 65, 80, 85, 90]
}
df = pd.DataFrame(data)

# 2. Scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Salary', 'Spending_Score']])

# 3. Perform hierarchical clustering
linked = linkage(scaled_features, method='ward')

# 4. Plot dendrogram with custom labels
plt.figure(figsize=(8, 5))
dendrogram(linked,
           labels=df['Customer_ID'].values,   # <-- use customer IDs instead of indices
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Customer Segmentation Dendrogram')
plt.xlabel('Customer')
plt.ylabel('Distance')
plt.show()

# 5. Choose number of clusters (e.g., 3)
df['Cluster'] = fcluster(linked, 3, criterion='maxclust')

# 6. Assign discount based on cluster
discount_map = {1: '10%', 2: '20%', 3: '30%'}
df['iPhone_Discount'] = df['Cluster'].map(discount_map)

print(df)
```
