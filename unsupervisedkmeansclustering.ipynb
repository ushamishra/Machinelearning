```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Sample dataset
data = {
    'Salary': [25000, 40000, 60000, 80000, 100000, 120000, 150000, 200000],
    'Spending_Score': [20, 35, 40, 60, 65, 80, 85, 90]
}
df = pd.DataFrame(data)

# 2. Scale features for better clustering
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Salary', 'Spending_Score']])

# 3. Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# 4. Assign discount based on cluster
discount_map = {0: '10%', 1: '20%', 2: '30%'}
df['iPhone_Discount'] = df['Cluster'].map(discount_map)

# 5. Visualize clusters
plt.scatter(df['Salary'], df['Spending_Score'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Salary')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation for iPhone Discount')
plt.show()

# 6. Display results
print(df)
```
